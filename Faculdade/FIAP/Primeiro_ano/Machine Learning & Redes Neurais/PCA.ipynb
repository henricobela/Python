```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_breast_cancer
```

```python
breast = load_breast_cancer()
```

```python
breast
```

```python
breast_data = breast.data
breast_labels = breast.target
```

```python
breast_data.shape
```

```python
breast_labels.shape
```

```python
labels = np.reshape(breast_labels, (569, 1))
```

```python
final = np.concatenate([breast_data, labels], axis=1)
```

```python
final.shape
```

```python
colunas = list(breast.feature_names)
colunas.append("label")
```

```python
df = pd.DataFrame(final, columns=colunas)
```

```python
df["label_categ"] = df["label"].apply(
    lambda x: "benigno" if x == 0 else "maligno"
)  # aplicacao desta funcao apenas pelo datavis
```

```python
df
```

```python
df["label_categ"].value_counts()
```

```python
del df["label"]
df
```

```python
from sklearn.preprocessing import StandardScaler as ss
```

```python
df.shape
```

```python
data = df.drop(columns=["label_categ"])
```

```python
scaler = ss()
df_scaled = scaler.fit_transform(data)
```

```python
from sklearn.decomposition import PCA
```

```python
n_components = 2
pca = PCA(n_components=n_components)
```

```python
pca_breast = pca.fit_transform(df_scaled)
```

```python
pca_breast
```

```python
df_pca = pd.DataFrame(
    pca_breast, columns=[str(i + 1) + "º CP" for i in range(n_components)]
)
```

```python
df_pca
```

```python
lista_pca = pca.explained_variance_ratio_
i = 0
for value in lista_pca:
    i += value

print(f"{i:.2%}")
```

```python
plt.figure(figsize=(10, 10))
plt.title("PCA do Dataset")
targets = ["benigno", "maligno"]
cores = ["g", "r"]

for t, c in zip(targets, cores):
    index = df["label_categ"] == t
    plt.scatter(df_pca.loc[index, "1º CP"], df_pca.loc[index, "2º CP"], c=c, label=t)

plt.legend()
plt.show
```
