```python
import random
```

```python
ix = [(row, col) for row in range(data.shape[0]) for col in range(data.shape[-1] - 1)]

for row, col in random.sample(ix, int(round(0.1 * len(ix)))):
    data.iat[row, col] = np.nan
```
