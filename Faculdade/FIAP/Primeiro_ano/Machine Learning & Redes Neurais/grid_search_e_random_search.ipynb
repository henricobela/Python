```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("http://tinyurl.com/42rvaw4p")
data
```

```python
data["smoker"] = data["smoker"].apply(lambda x: 1 if x == "yes" else 0)
data
```

```python
data["sex"] = data["sex"].apply(lambda x: 1 if x == "female" else 0)
data
```

```python
del data["region"]
```

```python
data
```

```python
from sklearn.model_selection import train_test_split

label = data["charges"]
del data["charges"]
x_train, x_test, y_train, y_test = train_test_split(data, label, test_size=0.30)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

ss.fit(x_train)

x_train = ss.transform(x_train)
x_test = ss.transform(x_test)
```

```python
n_estimators = [int(x) for x in np.linspace(100, 2000, num=20)]
max_features = ["sqrt", "auto"]
max_depth = [int(x) for x in np.linspace(5, 30, num=15)]
min_samples_split = [2, 5, 10, 20]
min_samples_leaf = [1, 2, 4, 5, 10]

params = {
    "n_estimators": n_estimators,
    "max_features": max_features,
    "max_depth": max_depth,
    "min_samples_split": min_samples_split,
    "min_samples_leaf": min_samples_leaf,
}
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(random_state=42)

grid_rf = GridSearchCV(estimator=rf, param_grid=params, cv=3, verbose=2, n_jobs=-1)
```

```python
# grid stoped
grid_rf.fit(x_train, y_train)
```

```python
random_rf = RandomizedSearchCV(
    estimator=rf,
    param_distributions=params,
    cv=3,
    verbose=2,
    n_jobs=-1,
    random_state=42,
    n_iter=100,
)
```

```python
random_rf.fit(x_train, y_train)
```

```python
random_rf.best_estimator_
```

```python
random_rf.best_score_
```

```python
best_model = random_rf.best_estimator_
best_model.fit(x_train, y_train)
y_pred = best_model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error as MSE
from sklearn.metrics import r2_score as r2
```

```python
print(f"MSE: {MSE(y_test, y_pred)}")
```

```python
print(f"R2: {round(r2(y_test, y_pred), 2)}")
```
