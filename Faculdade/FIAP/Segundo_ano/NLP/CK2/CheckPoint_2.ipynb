```python
from nltk.corpus import machado, stopwords
from nltk.text import Text
import spacy, re, nltk, string
import networkx as nx
from pyvis.network import Network


nltk.download('machado')
nltk.download("stopwords")
nltk.download("punkt")
nlp = spacy.load("pt_core_news_sm")
ptbr_stopwords = set(stopwords.words("portuguese"))
```

```python
texto = machado.raw('romance/marm05.txt')
```

```python
texto_sem_pontuacao = texto.translate(str.maketrans("", "", string.punctuation))
texto_sem_pontuacao = texto_sem_pontuacao.upper()
```

```python
palavras = nltk.word_tokenize(texto_sem_pontuacao)
palavras
```

```python
palavras_sem_stopwords = [palavra for palavra in palavras if palavra.lower() not in ptbr_stopwords]
palavras_sem_stopwords
```

```python
personagens_principais = [
                        'BRÁS CUBAS',
                        'VIRGÍLIA',
                        'CONSELHEIRO DUTRA',
                        'SABINA',
                        'MARCELA',
                        'EUGÊNIA',
                        'NHÃ-LOLÓ',
                        'LOBO NEVES',
                        'QUINCAS BORBA',
                        'DONA PLÁCIDA',
                        'PRUDÊNCIO',
                        'COTRIM']
```

```python
ocorrencias_personagens = []

for nome in personagens_principais:
    padrao = r'\b{0}\b'.format(re.escape(nome))
    ocorrencias = [palavras_sem_stopwords[i:i+20] for i in range(len(palavras_sem_stopwords)) if re.search(padrao, palavras_sem_stopwords[i])]
    ocorrencias_personagens.append((nome, ocorrencias))
    
```

```python
ocorrencias_personagens
```

```python
freq_coocorrencia = {}

for i in range(len(ocorrencias_personagens)):
    nome1, ocorrencias1 = ocorrencias_personagens[i]
    for j in range(i+1, len(ocorrencias_personagens)):
        nome2, ocorrencias2 = ocorrencias_personagens[j]
        freq = sum([1 for ocorrencia1 in ocorrencias1 for ocorrencia2 in ocorrencias2 if set(ocorrencia1) & set(ocorrencia2)])
        freq_coocorrencia[(nome1, nome2)] = freq
```

```python
freq_coocorrencia
```

```python
for nome, ocorrencias in ocorrencias_personagens:
    print(nome, ocorrencias)
```

```python
G = nx.Graph()

for personagem in personagens_principais:
    G.add_node(personagem)

for (nome1, nome2), freq in freq_coocorrencia.items():
    # G.add_edge(nome1, nome2, weight=freq)
    if freq > 0:
        G.add_edge(nome1, nome2, weight=freq)
        
net = Network(height='1000px', width='1000px', notebook=True)
net.from_nx(G)

for edge in net.edges:
    edge['width'] = '0.5'
    
net.show("graph.html")
```
