```python
import pyNN.spiNNaker as sim
import numpy as np
import matplotlib.pyplot as plt
```

```python
#setup da simulação
sim.setup(timestep=1.0, min_delay=1.0)
```

```python
#vamos criar um modelo neuronal
cell_params = {
    'cm' : 0.25, #0.25 nF a capacitancia da menbrana potencial Ok
    'tau_m': 10, #time de resposta numa mudança de voltagem ms Ok
    'tau_syn_E': 2, #periodo onde o neuronio não gera ação potencial Ok
    'tau_syn_I': 2.5, # 2.5 ms tempo de inibição Ok
    'v_reset' : -70.0, # -70 mv depois do acontecimento de um spike ok
    'v_rest' : -65.0, #o valor da tensão em repouso menbrana potencial Ok
    'v_thresh': -55.0 #valor limítrofe de disparo OK   
}
```

```python
#vamos criar uma população de 100 neuronios
neurons = sim.Population(100, sim.IF_cond_exp(**cell_params))
```

```python
neurons
```

```python
inputs = sim.Population(100,sim.SpikeSourcePoisson(rate=0.0) )
```

```python
input_firing_rates = np.linspace(0.0, 1000.0, num=inputs.size)
inputs.set(rate = input_firing_rates)
```

```python
wiring = sim.OneToOneConnector()
static_synapse = sim.StaticSynapse(weight=0.1, delay=2.0)
connections = sim.Projection(inputs, neurons, wiring, static_synapse)
```

```python
#salvando os spikes
neurons.record('spikes')
```

```python
sim.run(10000) #10 s
```

```python
spike_count = neurons.get_spike_counts()
spike_count
```

```python
output_firing_rate = np.array([value for (key,value) in sorted(spike_count.items())]) / 10
output_firing_rate
```

```python
plt.plot(input_firing_rates, output_firing_rate)
plt.savefig("spikes per rate")
```
