# Importando Libs

```python
from nltk.corpus import machado, stopwords
from nltk.text import Text
import spacy, re, nltk, string
import networkx as nx
from pyvis.network import Network
from nltk.util import ngrams
from collections import Counter


nltk.download('machado')
nltk.download("stopwords")
nltk.download("punkt")
nlp = spacy.load("pt_core_news_sm")
ptbr_stopwords = set(stopwords.words("portuguese"))
```

# Download do livro Memórias Póstumas de Brás Cubas de Machado de Assis

```python
texto = machado.raw('romance/marm05.txt')
```

# Limpeza do texto

```python
texto_sem_pontuacao = texto.translate(str.maketrans("", "", string.punctuation))
texto_sem_pontuacao = texto_sem_pontuacao.upper()
```

```python
palavras = nltk.word_tokenize(texto_sem_pontuacao)
palavras
```

```python
palavras_sem_stopwords = [palavra for palavra in palavras if palavra.lower() not in ptbr_stopwords]
palavras_sem_stopwords
```

```python
personagens_principais = [
                        'BRÁS CUBAS',
                        'VIRGÍLIA',
                        'CONSELHEIRO DUTRA',
                        'SABINA',
                        'MARCELA',
                        'EUGÊNIA',
                        'NHÃLOLÓ',
                        'LOBO NEVES',
                        'QUINCAS BORBA',
                        'DONA PLÁCIDA',
                        'PRUDÊNCIO',
                        'COTRIM']
```

```python
for idx, palavra in enumerate(palavras_sem_stopwords):
    for nome in personagens_principais:
        if palavra in nome:
            palavras_sem_stopwords[idx] = nome
```

```python
try:
    for idx, palavra in enumerate(palavras_sem_stopwords):
        if palavras_sem_stopwords[idx] == palavras_sem_stopwords[idx+1]:
            palavras_sem_stopwords.pop(idx+1)
except:
    pass
```

# Construção da Rede de relacionamento e pequeno bag of words

```python
def bow(tokens, n):
    n_grams = list(ngrams(tokens, n))
    return Counter(n_grams)

def rede(tokens, personagens):
    relacionamentos = {personagem: {} for personagem in personagens}

    ngram_counts = bow(tokens, 20)

    for personagem in personagens:
        for ngram, count in ngram_counts.items():
            if personagem in ngram:
                outros_personagens = [p for p in ngram if p != personagem and p in personagens]
                for outro_personagem in outros_personagens:
                    if outro_personagem not in relacionamentos[personagem]:
                        relacionamentos[personagem][outro_personagem] = count
                    else:
                        relacionamentos[personagem][outro_personagem] += count

    return relacionamentos

freq_coocorrencia = rede(palavras_sem_stopwords, personagens_principais)

for personagem, relacionamentos in freq_coocorrencia.items():
    print(f"Relacionamentos de {personagem} em um conjunto de 20 palavras:")
    for outro_personagem, count in relacionamentos.items():
        print(f"\t{outro_personagem}: {count}")
```

# View da Rede

```python
G = nx.Graph()
cores = ["#FF0000", "#00FF00", "#0000FF", "#FFFF00", "#FF00FF", "#00FFFF"] 

for i, personagem in enumerate(personagens_principais):
    cor = cores[i % len(cores)]
    G.add_node(personagem, size = 10, color = cor)

for personagem, relacionamentos in freq_coocorrencia.items():
    for outro_personagem, count in relacionamentos.items():
        G.add_edge(personagem, outro_personagem, weight = count, size = 10)
        
net = Network(height = '1200px', width = '100%', notebook = True,  directed = False)
net.barnes_hut(gravity = -8000, 
               central_gravity = 0.1, 
               spring_length = 100, 
               spring_strength = 0.001, 
               damping = 0.09, 
               overlap = 0.5)


for edge in net.edges:
    edge['width'] = '0.5'

for node in G.nodes:
    net.add_node(node, size = G.nodes[node]['size'], color = G.nodes[node]['color'])

for edge in G.edges:
    net.add_edge(edge[0], edge[1], width = 1)

net.show("graph.html")
```
