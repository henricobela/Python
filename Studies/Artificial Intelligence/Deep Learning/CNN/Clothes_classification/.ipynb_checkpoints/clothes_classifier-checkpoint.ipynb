```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import random

from sklearn.model_selection import train_test_split
from sklearn import metrics

print(f"TF Version: {tf.__version__}")
```

```python
df_train = pd.read_csv("fashion-mnist-train.csv", sep = ",")
df_test = pd.read_csv("fashion-mnist-test.csv", sep = ",")
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
training = np.array(df_train, dtype = "float32")
testing = np.array(df_test, dtype = "float32")
```

```python
training
```

```python
training.shape
```

```python
i = random.randint(1, 60000)
plt.imshow(training[i, 1:].reshape((28,28)))
plt.imshow(training[i, 1:].reshape((28,28)), cmap = "gray");

label = training[i,0]
label
```

```python
w_grid = 15
l_grid = 15

fig, axes = plt.subplots(l_grid, w_grid, figsize = (17,17))
axes = axes.ravel()
len_training = len(training)

for i in np.arange(0, w_grid * l_grid):
    index = np.random.randint(0, len_training)
    axes[i].imshow(training[index, 1:].reshape((28,28)))
    axes[i].set_title(training[index, 0], fontsize = 8)
    axes[i].axis("off")
plt.subplots_adjust(hspace = 0.4)
```

```python
x_train = training[:, 1:] / 255
y_train = training[:,0]

x_test = testing[:, 1:] / 255
y_test = testing[:,0]
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train = x_train.reshape(x_train.shape[0], *(28, 28, 1))
x_test = x_test.reshape(x_test.shape[0], *(28, 28, 1))
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
cnn = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation = "relu", input_shape = (28, 28, 1)),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(64, (3, 3), activation = "relu"),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(64, (3, 3), activation = "relu"),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation = "relu"),
    tf.keras.layers.Dense(10, activation = "softmax"), #retorna uma probabilidade para cada uma das classes    
])
```

```python
cnn.summary()
```

```python
cnn.compile(optimizer = "Adam", loss = "sparse_categorical_crossentropy", metrics = ["accuracy"])
```

```python
epochs = 15
history = cnn.fit(x_train, y_train, batch_size = 512, epochs = epochs) # Ajuste dos pesos ser de 512 em 512 registros
```

```python
evaluation = cnn.evaluate(x_test, y_test)
```

```python
predicted_classes = cnn.predict(x_test) 
classes_x = np.argmax(predicted_classes, axis = 1)
```

```python
classes_x
```

```python
y_test
```

```python
w_grid = 5
l_grid = 5

fig, axes = plt.subplots(l_grid, w_grid, figsize = (12,12))
axes = axes.ravel()

for i in np.arange(0, w_grid * l_grid):
    axes[i].imshow(x_test[i].reshape((28,28)))
    axes[i].set_title("Prediction class {:0.1f}\n True Class = {:0.1f}".format(classes_x[i], y_test[i]), fontsize = 8)
    axes[i].axis("off")
plt.subplots_adjust(hspace = 0.4)
```

```python
cm = metrics.confusion_matrix(y_test, classes_x)
plt.figure(figsize = (14,10))
sns.heatmap(cm, annot = True);
```

```python
cm
```

```python
num_class = 10
target_names = ["Class {}".format(i) for i in range(num_class)]
target_names
```

```python
print(f"{metrics.classification_report(y_test, classes_x, target_names = target_names)}")
```
