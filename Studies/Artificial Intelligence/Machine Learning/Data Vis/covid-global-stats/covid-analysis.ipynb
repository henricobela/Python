```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("dark")
```

```python
data_set_daily = pd.read_csv(r"data/worldometer_coronavirus_daily_data.csv")
data_set_summary = pd.read_csv(r"data/worldometer_coronavirus_summary_data.csv")
```

```python
df = data_set_daily
df
```

```python
df["continent"] = df.apply(
    lambda row: data_set_summary[data_set_summary.country == row.country]
    .iloc[0]
    .continent,
    axis=1,
)
```

```python
x = data_set_summary["total_deaths"].groupby(data_set_summary["continent"]).sum()
order = df["continent"].value_counts().index

values = np.arange(0, 2000000, 500000)
value_increment = 1

colors = ["r", "b", "g", "c", "k", "m"]

plt.figure(figsize=(7, 6))
plt.title("Contagem de Mortes por Continente")

ax = x.plot(kind="bar", color=colors)
# sns.countplot(x = x, hue = df["continent"], order = order, dodge = False)
# sns.countplot(x = df["continent"], hue = df["continent"], order = order, dodge = False)

plt.xticks(rotation=45)
plt.xlabel("Continentes")
plt.ylabel("Contagem")
plt.yticks(values * value_increment, ["%d" % val for val in values])

for patch in ax.patches:
    y_value = patch.get_height()
    x_value = patch.get_x() + patch.get_width() / 2
    space = 1
    label = format(y_value)
    ax.annotate(
        label,
        (x_value, y_value),
        xytext=(0, space),
        textcoords="offset points",
        ha="center",
        va="bottom",
    )

# plt.legend(["Africa", "Asia", "Oceania", "Europa", "North America", "South America"],loc = 2)

plt.show()
```
