# REDES NEURAIS - MODELOS

Utilizaremos o MLPRegressor, Decision Tree Regressor, RandomForest como nossos modelos de previsão

Importando recursos

```python
from sklearn.neural_network import MLPClassifier as MLPc
from sklearn.metrics import accuracy_score as acc
from sklearn.metrics import classification_report
```

Definindo as funções que serão utilizadas

```python
def plot(x, y, df):
    matrix(df, x, y)
    plt.title("Grafico de Matrix de Confusão")
    plt.xlabel("Previsões")
    plt.ylabel("Verdadeiros")
    plt.show()
```

```python
def metrics(y_test, y_pred):
    print("Metricas:")
    print(f"R2 Score: {r2_score(y_test, y_pred):.3f}")
```

```python
def mlp_model(x_train, x_test, y_train, y_test):
    mlp = MLPc(hidden_layer_sizes=(1,), max_iter=5000)

    mlp.fit(x_train, y_train)
    y_pred = mlp.predict(x_test)

    metrics(y_test, y_pred)
    plot(x_test, y_test, df=mlp)
```

```python
def mlp_model2(x_train, x_test, y_train, y_test):
    mlp = MLPc(hidden_layer_sizes=(3,), max_iter=10000)

    mlp.fit(x_train, y_train)
    y_pred = mlp.predict(x_test)

    metrics(y_pred, y_test)
    plot(x_test, y_test, df=mlp)
```

```python
def mlp_model3(x_train, x_test, y_train, y_test):
    mlp = MLPc(
        hidden_layer_sizes=(11,),
        random_state=4,
        solver="adam",
        beta_1=0.7,
        max_iter=8000,
        learning_rate_init=0.01,
        activation="relu",
    )

    mlp.fit(x_train, y_train)
    y_pred = mlp.predict(x_test)

    metrics(y_pred, y_test)
    plot(x_test, y_test, df=mlp)
```
