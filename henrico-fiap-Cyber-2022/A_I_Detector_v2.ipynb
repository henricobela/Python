```python
from fuzzywuzzy import process
```

```python
lista = ["quatro pernas", "cadeirante", "corre aí", "vamos andar", "ande", "se mexe", "muleta", "rodinhas", "rodinha", "G", "levanta", "sem pernas", "perninhas", "desrespeito", "não acolhimento", "desafeto", "perna", "anda"]
texto = input("Digite sua frase: ")
process.extract(texto, lista)
```

```python
def getMatch(field):
  match_list = ["quatro pernas", "cadeirante", "corre aí", "vamos andar", "ande", "se mexe", "muleta", "rodinhas", "rodinha", "G", "levanta", "sem pernas", "perninhas", "desrespeito", "não acolhimento", "desafeto", "perna", "anda"]
  search_list = process.extract(field, match_list)
  result = []

  for text in search_list:
    if text[1] > 54:
      result.append(text)
  
  if len(result) == 0:
    return "Não é Preconceito"
  return "Preconceito"

getMatch(input())
```
