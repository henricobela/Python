Detector de Preconceito para Pessoas com mobilidade reduzida


```python
import re
```

```python
#quatro pernas, cadeirante, corre aí, vamos andar, ande, se mexe, muleta, rodinhas, rodinha, G, levanta, sem pernas, perninhas, desrespeito, não acolhimento, desafeto
```

```python
lista = ["quatro pernas", "cadeirante", "corre aí", "vamos andar", "ande", "se mexe", "muleta", "rodinhas", "rodinha", "G", "levanta", "sem pernas", "perninhas", "desrespeito", "não acolhimento", "desafeto"]
```

```python
texto = input() #programa pede para digitar algo, e reserva na variavel texto
if texto in lista: #se a coisa foi digitada estiver na lista
  print("Desrespeito, não inclusão, Racismo") #mostra na tela #quanto tem um espaço é chamado de identação, quando se aperta a tecla TAB
else: #caso contrario
  print("Voce é muito legal!!!") #mostra na tela
```

```python
def getMatchRegex(field):
  match = re.search('(?i)(quatro pernas)|(perna)|(andar)|(perneta)|(preconceito)|(cadeirante)|(corre aí)|(vamos andar)|(ande)|(se mexe)|(muleta)|(rodinhas)|(rodinha)|(cadeira de rodas)|(levanta)|(sem pernas)|(perninhas)|(desrespeito)|(não acolhimento)|(desafeto)|(roda)', field.lower())

  if match:
    print("Desrespeito, não inclusão, Racismo")
  else:
    print("Voce é muito legal!!!")

getMatchRegex(input())
```
