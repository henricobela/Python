```python
# importando a biblioteca para conexao do nosso sistema na web
from pyngrok import ngrok
import streamlit as st
#import streamlit_ace as st_ace
import re
```

```python
#Comando para criar o sistema no arquivo app.py
%%writefile app.py

# essa funcao é para criacao da webpage
def main():
  #codigo para elementos da web - em HTML
  html_temp = """
  <div style ="background-color:red;padding:13px">
  <h1 style ="color:white;text-align:center;">iFood</h1>
  </div>
  """
  st.markdown(html_temp, unsafe_allow_html = True)

  titulo = st.text_input('Título do anúncio')
```

```python
def getMatchRegex(field):
  match = re.search('(?i)(quatro pernas)|(perna)|(andar)|(perneta)|(preconceito)|(cadeirante)|(corre aí)|(vamos andar)|(ande)|(se mexe)|(muleta)|(rodinhas)|(rodinha)|(cadeira de rodas)|(levanta)|(sem pernas)|(perninhas)|(desrespeito)|(não acolhimento)|(desafeto)|(roda)', field.lower())
  if st.button("Verificar"):
    if match:
      st.success("É Preconceito!")
    else:
      st.success("Não é Preconceito!")

  getMatchRegex(input())
  
  if __name__=='__main__': 
    main()  
```
