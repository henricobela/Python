```python
import json
import requests
```

```python
res = requests.get("https://henricobela.github.io/fra.json")
```

```python
def getScoreTeste(field):
  dict_fra = res.json()["fra"]
  for item in dict_fra:
    for value in item.values():
      if str(field) == str(value):
        return "Não pode entrar no país"
      return "Pode entrar no país"
getScoreTeste(input())
```
