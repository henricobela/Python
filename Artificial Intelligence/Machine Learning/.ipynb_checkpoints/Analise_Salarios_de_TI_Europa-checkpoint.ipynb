```python
df = df.rename(columns = {"Have you received additional monetary support from your employer due to Work From Home? If yes, how much in 2020 in EUR": "Additional_Monetary"})
df = df.rename(columns = {"Have you been forced to have a shorter working week (Kurzarbeit)? If yes, how many hours per week": "Forced_work_hours"})
df = df.rename(columns = {"Have you lost your job due to the coronavirus outbreak?": "Loss_job_covid"})
df = df.rename(columns = {"Number of vacation days": "Vacation"})
df = df.rename(columns = {"Employment status": "Employment"})
df = df.rename(columns = {"Сontract duration": "Contract_duration"})
df = df.rename(columns = {"Annual bonus+stocks one year ago. Only answer if staying in same country": "Annual_stocks"})
df = df.rename(columns = {"Annual brutto salary (without bonus and stocks) one year ago. Only answer if staying in the same country": "Annual_brutto_Salary"})
df = df.rename(columns = {"Yearly bonus + stocks in EUR": "Year_stock_bonus"})
df = df.rename(columns = {"Yearly brutto salary (without bonus and stocks) in EUR": "Brutto_salary_no_bonus"})
df = df.rename(columns = {"Other technologies/programming languages you use often": "Programming_languages_useless"})
df = df.rename(columns = {"Your main technology / programming language": "Main_programming_language"})
df = df.rename(columns = {"Years of experience in Germany": "Year_exp_german"})
df = df.rename(columns = {"Total years of experience": "Total_exp"})
df = df.rename(columns = {"Are you getting any Stock Options?": "Stock_opt"})
df = df.rename(columns = {"Salary two years ago": "Salary_2_year_ago"})
df = df.rename(columns = {"Salary one year ago": "Salary_1_year_ago"})
df = df.rename(columns = {"Current Salary": "Salary"})
df = df.rename(columns = {"Your level": "Level"})
df = df.rename(columns = {"Years of experience": "Exp"})
df = df.rename(columns = {"Main language at work": "Language"})
```

```python
plt.figure(figsize = (8, 6))

sns.heatmap(df.corr(), annot = True)

plt.title("Grafico de Correlação Winner x Features", fontsize = 20)

plt.show()
```

```python
transformer = lambda x: x.astype("category").cat.codes

#for coluna in df.columns:
#    if coluna == "Salary":
#        pass
#    else:
#        df[coluna] = transformer(df[coluna])

df["Timestamp"] = transformer(df["Timestamp"])
df["Gender"] = transformer(df["Gender"])
df["City"] = transformer(df["City"])
df["Position"] = transformer(df["Position"])
df["Level"] = transformer(df["Level"])
df["Stock_opt"] = transformer(df["Stock_opt"])
df["Language"] = transformer(df["Language"])
df["Company size"] = transformer(df["Company size"])
df["Company type"] = transformer(df["Company type"])
df["Total_exp"] = transformer(df["Total_exp"])
df["Year_exp_german"] = transformer(df["Year_exp_german"])
df["Seniority level"] = transformer(df["Seniority level"])
df["Main_programming_language"] = transformer(df["Main_programming_language"])
df["Programming_languages_useless"] = transformer(df["Programming_languages_useless"])
df["Year_stock_bonus"] = transformer(df["Year_stock_bonus"])
df["Annual_stocks"] = transformer(df["Annual_stocks"])
df["Vacation"] = transformer(df["Vacation"])
df["Employment"] = transformer(df["Employment"])
df["Contract_duration"] = transformer(df["Contract_duration"])
df["Loss_job_covid"] = transformer(df["Loss_job_covid"])
df["Additional_Monetary"] = transformer(df["Additional_Monetary"])
```

```python
df.info()
```

```python
for coluna in df.columns:
    moda = statistics.mode(df[coluna])
    if df[coluna].dtype == "float64":
        df[coluna] = df[coluna].fillna(value = df[coluna].mean())
    if df[coluna].dtype == "int64":
        df[coluna] = df[coluna].fillna(value = df[coluna].mean())
    if df[coluna].dtype == "O":
        df[coluna] = df[coluna].fillna(value = moda)
```

```python
from sklearn.preprocessing import StandardScaler as ss
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression as LR
from sklearn.metrics import r2_score
```

```python
X = df.drop(columns = ["Salary", "Main_programming_language", "Total_exp","Year_exp_german", "Vacation",
                       "Seniority level", "Employment", "Contract_duration", "Loss_job_covid", 
                       "Timestamp", "Company type", "Language", "Programming_languages_useless"])
y = df["Salary"]

x_train, x_test, y_train, y_test = tts(X, y,
                                      test_size = 0.2,
                                      random_state = 42)
```

```python
regressor = LR()
regressor.fit(x_train, y_train)

y_pred = regressor.predict(x_test)

r2 = r2_score(y_test, y_pred)

df_temp = pd.DataFrame({"Real": y_test, "Predict": y_pred})
df_temp.sample(5)

print(f"{r2:.3f}")
display(df_temp.sample(5))
```

```python
scaler = ss()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
regressor = LR()
regressor.fit(x_train_scaled, y_train)

y_pred = regressor.predict(x_test_scaled)

r2 = r2_score(y_test, y_pred)

df_temp = pd.DataFrame({"Real": y_test, "Predict": y_pred})
df_temp.sample(5)

print(f"{r2:.3f}")
display(df_temp.sample(5))
```
