```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_set_daily = pd.read_csv(r"data/worldometer_coronavirus_daily_data.csv")
data_set_summary = pd.read_csv(r"data/worldometer_coronavirus_summary_data.csv")
```

```python
data_set_daily
```

```python
data_set_summary
```

```python
df = data_set_daily
df
```

```python
df['continent'] = df.apply(lambda row: data_set_summary[data_set_summary.country == row.country].iloc[0].continent, axis = 1)
```

```python
df
```

```python
df.columns
```

```python
for coluna in df.columns:
    count = df[coluna].value_counts()
    print(f"Coluna: {coluna}\n\nContagens:\n{count}\n\n\n\n")
```

```python
df.isnull().sum()
```

```python
data_set_summary["total_deaths"].groupby(df["continent"])
```

```python
continent_count = df.continent.value_counts()
total_deaths = data_set_summary["total_deaths"].groupby(df["continent"])

plt.figure(figsize = (7, 7))
plt.title("Count Plot by Continent")

sns.countplot(x = total_deaths, hue = df["continent"])

plt.xticks(df["continent"])

plt.show()
```
